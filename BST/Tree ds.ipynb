# Binary Search Tree (BST) Comprehensive Guide

## 1. What is a Binary Search Tree (BST)?
A Binary Search Tree (BST) is a type of binary tree where each node has the following properties:
- **Left Subtree:** Contains only nodes with values less than the node’s key.
- **Right Subtree:** Contains only nodes with values greater than the node’s key.
- **No Duplicates:** No two nodes have the same key.

## 2. Explain the properties of a Binary Search Tree.
The main properties of a BST are:

| Property                | Description                                                                 |
|-------------------------|-----------------------------------------------------------------------------|
| Left Subtree            | Nodes have values less than the parent node.                                |
| Right Subtree           | Nodes have values greater than the parent node.                             |
| Unique Values           | Each node’s value is unique.                                                |
| Recursive Structure     | Each subtree is itself a BST.                                               |

## 3. How does a BST differ from a binary tree?
| Feature                | Binary Tree                          | Binary Search Tree (BST)             |
|------------------------|--------------------------------------|--------------------------------------|
| Node Values            | No specific order                    | Left < Root < Right                  |
| Searching Efficiency   | O(n) in the worst case               | O(log n) in the average case         |
| Duplicates             | Allowed                              | Typically not allowed                |

## 4. What are the applications of BSTs in real-world scenarios?
- **Database Indexing:** Efficiently search and retrieve data.
- **Network Routing Algorithms:** Quickly find the shortest path.
- **Autocomplete Systems:** Quickly suggest words based on input.

## 5. Why is a BST considered an efficient data structure?
- **Balanced Trees:** Operations like search, insert, and delete take O(log n) time in a balanced BST.
- **Sorted Order:** Inorder traversal gives elements in sorted order.
- **Dynamic:** Easier to manage than arrays when dealing with dynamic data.

## 6. How do you insert a node into a BST?
1. Start at the root.
2. Compare the node’s value with the root.
3. If smaller, move to the left; if larger, move to the right.
4. Repeat until the correct position is found.
5. Insert the node as a leaf.

## 7. Explain the process of searching for a value in a BST.
1. Start at the root.
2. Compare the target value with the current node’s value.
3. Move left if the target is smaller, right if it’s larger.
4. Repeat until the value is found or a leaf is reached.

## 8. Describe the steps to delete a node from a BST.
1. **Leaf Node:** Simply remove it.
2. **Node with One Child:** Replace the node with its child.
3. **Node with Two Children:** Replace with the in-order predecessor (maximum in the left subtree) or successor (minimum in the right subtree).

## 9. What are the three cases to consider when deleting a node in a BST?
- **Case 1:** Node is a leaf.
- **Case 2:** Node has one child.
- **Case 3:** Node has two children.

## 10. How do you handle the deletion of a node with two children in a BST?
- Find the in-order predecessor or successor.
- Replace the node’s value with the predecessor or successor’s value.
- Delete the predecessor or successor, which will now have at most one child.

## 11. What is an inorder, preorder, and postorder traversal in a BST?
| Traversal  | Order of Nodes Processed                  |
|------------|-------------------------------------------|
| Inorder    | Left -> Root -> Right                     |
| Preorder   | Root -> Left -> Right                     |
| Postorder  | Left -> Right -> Root                     |

- **Inorder:** Yields nodes in ascending order for a BST.

## 12. How can you print all elements in a BST in sorted order?
Perform an **Inorder Traversal**:
1. Traverse the left subtree.
2. Visit the root.
3. Traverse the right subtree.

## 13. Explain how to find the minimum and maximum values in a BST.
- **Minimum Value:** Traverse to the leftmost node.
- **Maximum Value:** Traverse to the rightmost node.

## 14. Describe how to find the height of a BST.
- **Height:** The number of edges on the longest path from the root to a leaf.
- **Formula:** `Height = 1 + max(Height of Left Subtree, Height of Right Subtree)`

## 15. How do you find the predecessor and successor of a node in a BST?
- **Predecessor:** The maximum value in the left subtree.
- **Successor:** The minimum value in the right subtree.

## 16. What is the time complexity of search, insertion, and deletion operations in a BST?
| Operation | Average Case Time Complexity | Worst Case Time Complexity |
|-----------|------------------------------|----------------------------|
| Search    | O(log n)                      | O(n)                        |
| Insertion | O(log n)                      | O(n)                        |
| Deletion  | O(log n)                      | O(n)                        |

## 17. Explain how balancing a BST improves its performance.
- **Balanced Tree:** Ensures O(log n) time complexity for operations.
- **Unbalanced Tree:** Can degrade to O(n) (e.g., when elements are inserted in sorted order).

## 18. What is a self-balancing BST? Give examples.
- **Self-Balancing BST:** Automatically maintains balance after insertions and deletions.
- **Examples:** AVL Tree, Red-Black Tree, Splay Tree.

## 19. Compare and contrast AVL Trees and Red-Black Trees.
| Feature          | AVL Tree                          | Red-Black Tree                     |
|------------------|-----------------------------------|------------------------------------|
| Balance Factor   | Strictly balanced (height difference ≤ 1) | Less strict (allows more imbalance) |
| Rotations        | More frequent (on every insert/delete) | Less frequent                      |
| Performance      | Faster lookups                    | Faster inserts and deletes         |

## 20. How does an AVL tree maintain its balance?
- By **Rotations:** Single or double rotations to restore balance after insertions or deletions.
  
## 21. What are rotations in AVL Trees? Explain with examples.
- **Left Rotation:** Rotate a subtree right-heavy to the left.
- **Right Rotation:** Rotate a subtree left-heavy to the right.
- **Double Rotation:** Combination of left and right rotations (e.g., Left-Right or Right-Left).

## 22. Describe the role of color properties in Red-Black Trees.
- **Red and Black:** Nodes are colored to ensure the tree remains balanced.
- **Properties:** 
  - Root is black.
  - Red nodes cannot have red children (no consecutive red nodes).
  - Every path from a node to its descendants has the same number of black nodes.

## 23. How do you convert a BST to a Doubly Linked List (DLL)?
1. Perform an **Inorder Traversal**.
2. Modify pointers to link nodes as a DLL instead of using tree links.

## 24. What is the difference between a BST and a Heap?
| Feature       | Binary Search Tree (BST)           | Heap                                 |
|---------------|------------------------------------|--------------------------------------|
| Order Property| Left < Root < Right                | Parent is greater/smaller than children|
| Use Case      | Searching and Sorting              | Priority Queue                       |

## 25. How can you merge two BSTs?

2. Merge the two sorted arrays.
3. Convert the merged array back to a BST.

## 26. How do you check if a given binary tree is a BST?
1. Traverse the tree using **Inorder Traversal**.
2. Ensure the resulting sequence is in ascending order.

## 27. Write a function to find the Lowest Common Ancestor (LCA) in a BST.
```python
def find_LCA(root, n1, n2):
    while root:
        if root.data > n1 and root.data > n2:
            root = root.left
        elif root.data < n1 and root.data < n2:
            root = root.right
        else:
            break
    return root
```

## 31. Write an algorithm to find the diameter of a BST.
- **Diameter:** The longest path between any two nodes in the BST.
- **Algorithm:**
  1. Compute the height of left and right subtrees.
  2. Diameter = max(left height + right height + 1, left diameter, right diameter).

## 32. How do you find the distance between two nodes

 in a BST?

2. Find the distance from LCA to each node.
3. Distance between nodes = Distance to first node + Distance to second node.

## 33. How can you find the sum of all elements in a BST?

## 34. What is the Morris Traversal, and how does it work for inorder traversal of a BST?
- **Morris Traversal:** A way to traverse the BST without recursion or a stack, using threaded binary trees.
- **Steps:**
  1. If left child is null, visit the node.
  2. If left child is not null, find the inorder predecessor and make the current node the right child of the predecessor.
  3. Restore the tree structure after visiting.

## 35. Explain the process to flatten a BST to a linked list in-place.
- Use **Right Skewed Transformation:**
  1. Traverse the tree using **Inorder Traversal**.
  2. For each node, move the left child to the right and set the left pointer to null.

## 36. What is the impact of inserting elements in a sorted order into a BST?

## 37. How do degenerate or skewed BSTs affect performance?
- **Degenerate Trees** behave like linked lists, degrading search, insert, and delete operations to O(n).

## 38. What is the relationship between BSTs and Binary Heaps?

## 39. Can a BST have duplicate elements? How can duplicates be handled?
- **Duplicates:** Typically not allowed in standard BSTs.
- **Handling Duplicates:** Store duplicates in a secondary structure, or allow duplicates on either left or right subtree based on specific rules.

## 40. How do you implement a BST in a non-recursive manner?

## 41. What are threaded binary trees? How are they different from standard BSTs?
- **Threaded Binary Trees:** Special type of binary tree where null pointers are replaced with pointers to the inorder predecessor or successor.
- **Difference:** Threaded trees optimize traversal without using stacks or recursion.

## 42. Explain the concept of a Treap (Tree + Heap). How does it differ from a standard BST?
- **Treap:** Combines properties of a BST and a Heap.
  - **BST Property:** Based on node keys.
  - **Heap Property:** Based on node priorities.

## 43. What is the Cartesian Tree, and how does it relate to BSTs?
- **Cartesian Tree:** A binary tree derived from a sequence where:
  - **Inorder Traversal:** Yields the sequence.
  - **Heap Property:** The parent node is smaller than its children.
- **Relation to BSTs:** Maintains both BST and Heap properties.

## 44. Discuss the concept of a Splay Tree and its applications.
- **Splay Tree:** A self-adjusting BST where recently accessed elements are moved to the root.
- **Applications:** Fast access to frequently accessed elements (e.g., caches, data compression).

## 45. What is an order statistic tree? How is it implemented using a BST?
- **Order Statistic Tree:** A type of BST that supports efficient order-based operations (e.g., finding the kth smallest element).

## 46. What is the average case time complexity for searching in a BST?
- **Average Case:** O(log n) when the tree is balanced.

## 47. Why does a BST perform poorly on already sorted data?

## 48. How does tree rotation affect the time complexity of operations in a BST?

````python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = TreeNode(val)
        else:
            self._insert(self.root, val)

    def _insert(self, node, val):
        if val < node.val:
            if node.left is None:
                node.left = TreeNode(val)
            else:
                self._insert(node.left, val)
        else:
            if node.right is None:
                node.right = TreeNode(val)
            else:
                self._insert(node.right, val)

    def inorder_traversal(self):
        return self._inorder_traversal(self.root, [])

    def _inorder_traversal(self, node, result):
        if node:
            self._inorder_traversal(node.left, result)
            result.append(node.val)
            self._inorder_traversal(node.right, result)
        return result

    def preorder_traversal(self):
        return self._preorder_traversal(self.root, [])

    def _preorder_traversal(self, node, result):
        if node:
            result.append(node.val)
            self._preorder_traversal(node.left, result)
            self._preorder_traversal(node.right, result)
        return result

    def postorder_traversal(self):
        return self._postorder_traversal(self.root, [])

    def _postorder_traversal(self, node, result):
        if node:
            self._postorder_traversal(node.left, result)
            self._postorder_traversal(node.right, result)
            result.append(node.val)
        return result
````

````python
# Create a binary tree
bt = BinaryTree()

# Insert nodes into the tree
bt.insert(50)
bt.insert(30)
bt.insert(70)
bt.insert(20)
bt.insert(40)
bt.insert(60)
bt.insert(80)

# Perform different traversals
print("Inorder Traversal:", bt.inorder_traversal())  # Output: [20, 30, 40, 50, 60, 70, 80]
print("Preorder Traversal:", bt.preorder_traversal()) # Output: [50, 30, 20, 40, 70, 60, 80]
print("Postorder Traversal:", bt.postorder_traversal()) # Output: [20, 40, 30, 60, 80, 70, 50]
````

````python
import matplotlib.pyplot as plt
import networkx as nx

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = TreeNode(val)
        else:
            self._insert(self.root, val)

    def _insert(self, node, val):
        if val < node.val:
            if node.left is None:
                node.left = TreeNode(val)
            else:
                self._insert(node.left, val)
        else:
            if node.right is None:
                node.right = TreeNode(val)
            else:
                self._insert(node.right, val)

    def plot_tree(self):
        G = nx.DiGraph()
        self._add_edges(self.root, G)
        pos = self._get_positions(self.root, 0, 0, 2 ** self._get_depth(self.root))
        nx.draw(G, pos, with_labels=True, arrows=True, node_size=2000, node_color='lightblue', font_size=10, font_weight='bold')
        plt.show()

    def _add_edges(self, node, graph, parent=None):
        if node is not None:
            if parent:
                graph.add_edge(parent.val, node.val)
            self._add_edges(node.left, graph, node)
            self._add_edges(node.right, graph, node)

    def _get_positions(self, node, x, y, dx):
        pos = {}
        if node:
            pos[node.val] = (x, y)
            if node.left:
                pos.update(self._get_positions(node.left, x - dx, y - 1, dx / 2))
            if node.right:
                pos.update(self._get_positions(node.right, x + dx, y - 1, dx / 2))
        return pos

    def _get_depth(self, node):
        if node is None:
            return 0
        else:
            left_depth = self._get_depth(node.left)
            right_depth = self._get_depth(node.right)
            return max(left_depth, right_depth) + 1
````

````python
import matplotlib.pyplot as plt
import numpy as np

def plot_tree_from_array(arr):
    if not arr:
        print("Tree is empty.")
        return
    
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.axis('off')
    
    # Calculate depth of the tree
    depth = int(np.log2(len(arr))) + 1
    max_width = 2 ** depth

    def plot_node(index, x, y, dx):
        if index >= len(arr):
            return
        
        # Plot the node
        ax.text(x, y, str(arr[index]), fontsize=12, ha='center', va='center',
                bbox=dict(facecolor='lightblue', edgecolor='black', boxstyle='round,pad=0.3'))
        
        # Calculate children positions
        left_index = 2 * index + 1
        right_index = 2 * index + 2
        next_y = y - 1
        next_dx = dx / 2
        
        if left_index < len(arr):
            ax.plot([x, x - next_dx], [y, next_y], color='black')
            plot_node(left_index, x - next_dx, next_y, next_dx)
        if right_index < len(arr):
            ax.plot([x, x + next_dx], [y, next_y], color='black')
            plot_node(right_index, x + next_dx, next_y, next_dx)

    # Start plotting from root
    plot_node(0, 0, 0, max_width / 2)
    
    plt.show()

# Example usage
# Array representation of a binary tree
# Binary tree: 
#         1
#       /   \
#      2     3
#     / \   / \
#    4   5 6   7

tree_array = [1, 2, 3, 4, 5, 6, 7,8,7,5,43,2]
plot_tree_from_array(tree_array)
````

````python
class Node:
    def __init__(self,val) -> None:
        self.left=None
        self.right=None
        self.val=val


def inseart(root,val):
    if not root:
        return Node(val)
    else:
        if root.val==val:
            return root
        elif root.val<val:
            root.right=inseart(root.right,val)
        else:
            root.left=inseart(root.left,val)
    return root

def inOrder(root):
    if root:
        inOrder(root.left)
        print(root.val, end=' ')
        inOrder(root.right)

def search(root, key):
    if root is None:
        return None
    if root.val == key:
        return root
    elif root.val < key:
        return search(root.right, key)
    else:
        return search(root.left, key)
    
def LevelOrder(root):
    if not root:
        return
    queue=deque([root])
    while queue:
        current=queue.popleft()
        print(current.val, end=' ')
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
        
    
def search_levele(root, key, level=0):
    if root is None:
        return None  
    if root.val == key:
        return level 
    elif root.val < key:
        return search_levele(root.right, key, level + 1) 
    else:
        return search_levele(root.left, key, level + 1)  
    
def minNode(root):
    current = root
    while current and current.left:
        current = current.left
    return current

def maxNode(root):
    current = root
    while current and current.right:
        current = current.right
    return current

def delete_ele(root, ele):
    if not root:
        return root

    if ele < root.val:
        root.left = delete_ele(root.left, ele)
    elif ele > root.val:
        root.right = delete_ele(root.right, ele)
    else:
        # Node with only one child or no child
        if not root.left:
            return root.right
        elif not root.right:
            return root.left

        # Node with two children: Get the inorder successor
        tempNode = minNode(root.right)
        root.val = tempNode.val
        root.right = delete_ele(root.right, tempNode.val)

    return root



def find_InOrder_predeseccor(root):
    current=root
    while current.left:
        current=current.left
    return current.val
````

````python
##LevelOrder
def LevelOrder(root):
    if not root: return 
    Q=deque([root])
    res=[]
    while Q:
        Level=[]
        for _ in range(len(Q)):
            curr=Q.popleft()
            Level.append(curr.val)

            if curr.left:
                Q.append(curr.left)
            if curr.right:
                Q.append(curr.right)
        res.append(Level)
    return res
````
